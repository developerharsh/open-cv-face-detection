```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#reading image
test1 = cv2.imread('image.jpeg')
#converting image from BGR to gray
gray_img = cv2.cvtColor(test1, cv2.COLOR_BGR2GRAY)
```

```python
#displaying greyscale image
plt.imshow(gray_img, cmap='gray')
```

```python
#applying haar classifier
haar_face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_alt.xml')
```

```python
#detecting multiple faces in an image
faces = haar_face_cascade.detectMultiScale(gray_img, scaleFactor=1.1, minNeighbors=5)
print('Faces found: ', len(faces))
```

```python
#drawing rectangle around face
for (x, y, w, h) in faces:
    cv2.rectangle(test1, (x, y), (x+w, y+h), (0, 255, 0), 2)
#displaying image after converting to rgb
test1=cv2.cvtColor(test1, cv2.COLOR_BGR2RGB)
plt.imshow(test1)
```
